```python
import whisper
import json

model = whisper.load_model("base")

result = model.transcribe(audio_path)


print("\n--- Transcription ---")
print(result["text"])


print("\n--- Transcription with Potential Speaker Segments ---")
segments = result["segments"]
speaker_turn = 0 # Initialize speaker turn counter
for segment in segments:
  start = segment['start']
  end = segment['end']
  text = segment['text'].strip()

  # Manually assign speaker labels based on turns
  speaker_label = "doctor" if speaker_turn % 2 == 0 else "patient"
  speaker_turn += 1 # Increment speaker turn counter


  print(f"[{start:.1f}s - {end:.1f}s] ({speaker_label}): {text}")
```

```python
# Function to find the speaker for a given time segment based on diarization
def find_speaker(segment_start, segment_end, diarization_result):
    for turn, _, speaker in diarization_result.itertracks(yield_label=True):
        # Check for overlap between the transcription segment and diarization turn
        if max(segment_start, turn.start) < min(segment_end, turn.end):
            return speaker
    return "Unknown Speaker" # Return "Unknown Speaker" if no overlap is found

print("\n--- Transcription with Diarization-based Speaker Segments ---")
# Iterate through transcription segments and assign speaker based on diarization
for segment in segments:
    start = segment['start']
    end = segment['end']
    text = segment['text'].strip()

    speaker_label = find_speaker(start, end, diarization)

    print(f"[{start:.1f}s - {end:.1f}s] ({speaker_label}): {text}")
```

```python
# STEP 1: Install Required Packages
!pip install pandas fuzzywuzzy[speedup] --quiet

# STEP 2: Import Libraries
import pandas as pd
from fuzzywuzzy import process

# STEP 3: Load the ICD-10 and CPT files
try:
    icd_df = pd.read_csv('/content/ICD-10 (1).csv', encoding='latin-1', encoding_errors='ignore')  # Adjust path if needed
    print("✅ Loaded ICD-10 data.")
except Exception as e:
    print(f"Error loading ICD-10 file: {e}")
    icd_df = None # Ensure df is None if loading fails

try:
    cpt_df = pd.read_csv('/content/CPT (1).csv', encoding='latin-1', encoding_errors='ignore')
    print("✅ Loaded CPT data.")
except Exception as e:
    print(f"Error loading CPT file: {e}")
    cpt_df = None # Ensure df is None if loading fails


# STEP 4: Explore & Clean Columns (Update based on your actual CSV headers)
if icd_df is not None:
    print("ICD-10 Columns:", icd_df.columns.tolist())
if cpt_df is not None:
    print("CPT Columns:", cpt_df.columns.tolist())

# Rename for consistency (adjust as needed) - only if dataframes loaded successfully
if icd_df is not None and not icd_df.empty:
    # Assuming the first column is description and the second is code based on previous head()
    icd_df.columns = ['ICD_Description'] + [f'ICD_Code_{i}' for i in range(1, len(icd_df.columns))]
    # Let's try to use the first 'Unnamed' column after description as the code column
    icd_code_column_name = None
    for col in icd_df.columns:
        if 'ICD_Code_' in col:
            icd_code_column_name = col
            break
    if icd_code_column_name:
         icd_df = icd_df.rename(columns={icd_code_column_name: 'ICD_Code'})
    else:
         print("Warning: Could not find a suitable ICD code column to rename.")


if cpt_df is not None and not cpt_df.empty:
     # Assuming the first column contains both description and code based on previous head()
     cpt_df.columns = ['CPT_Data'] + [f'CPT_Extra_{i}' for i in range(1, len(cpt_df.columns))]
     # For CPT, we might need a different approach to separate code and description if they are in the same column


# Optional: convert everything to lowercase for better matching
if icd_df is not None and 'ICD_Description' in icd_df.columns:
    icd_df['ICD_Description'] = icd_df['ICD_Description'].astype(str).str.lower()

# For CPT, we'll work with the combined 'CPT_Data' for now
if cpt_df is not None and 'CPT_Data' in cpt_df.columns:
    cpt_df['CPT_Data'] = cpt_df['CPT_Data'].astype(str).str.lower()


# STEP 5: Function to search ICD & CPT by symptom
def get_codes_from_symptom(symptom, top_n=3):
    symptom = str(symptom).lower()

    print(f"\n🔍 Symptom: {symptom.title()}")

    # Search ICD-10
    print("\n🩺 ICD-10 Code Matches:")
    if icd_df is not None and 'ICD_Description' in icd_df.columns and 'ICD_Code' in icd_df.columns:
        # Ensure we are working with string types for fuzzy matching
        choices = icd_df['ICD_Description'].dropna().tolist()
        icd_matches = process.extract(symptom, choices, limit=top_n)

        for desc, score in icd_matches:
            # Find the corresponding code(s) for the matched description
            codes = icd_df[icd_df['ICD_Description'] == desc]['ICD_Code'].dropna().tolist()
            if codes:
                 # Join multiple codes if they exist for the same description
                 print(f"→ {', '.join(map(str, codes))} | {desc} (score: {score})")
            else:
                 print(f"→ No code found | {desc} (score: {score})")

    else:
        print("  ICD-10 data not available or columns not found for matching.")


    # Search CPT
    print("\n🛠️ CPT Code Matches:")
    if cpt_df is not None and 'CPT_Data' in cpt_df.columns:
        # For CPT, we are searching within the 'CPT_Data' column which contains both description and potential codes
        choices = cpt_df['CPT_Data'].dropna().tolist()
        cpt_matches = process.extract(symptom, choices, limit=top_n)

        for data, score in cpt_matches:
            # In a real scenario, you would need to parse the 'data' string to extract the CPT code
            # For this example, we'll just print the matching string
            print(f"→ {data} (score: {score})")
    else:
        print("  CPT data not available or 'CPT_Data' column not found for matching.")


# STEP 6: Test It Out
# Use the heuristically identified medical terms from the previous step (assuming they are still in memory)
if 'identified_symptoms' in locals() and 'identified_diseases' in locals() and 'identified_procedures' in locals():
    all_medical_terms = identified_symptoms + identified_diseases + identified_procedures
    print("\n--- Attempting to find codes for extracted medical terms ---")
    for term in all_medical_terms:
        get_codes_from_symptom(term)
else:
    print("\nCould not find 'identified_symptoms', 'identified_diseases', or 'identified_procedures'. Please run the NLP extraction step first.")
```
